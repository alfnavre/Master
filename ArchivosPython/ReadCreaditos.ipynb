```python
# Lectura de ficheros
rutaC = "C:\\Users\\CarmenGonzalezConde\\Documents\\MASTER\\"
rutaA = ""
rutaO = ""

ruta = rutaC + "PRACTICA_PYTHON-master\\PRACTICA_PYTHON-master\\BaseDeDatos\\"
file = ruta + "tmdb_5000_credits.csv"
file2 = ruta + "tmdb_5000_movies.csv"

credits = pd.read_csv(file, encoding="cp437", delimiter = ',')
movies = pd.read_csv(file2, encoding="cp437", delimiter = ',')
```

```python
# Merge de Dataframes 

# Cambiamos el nombre a la columa del dataframe movies para poder hacer el merge.
movies = movies.rename(columns= {'id': 'movie_id'})

# Hacemos el merge
dfMerge = pd.merge(credits, movies, on='movie_id')

# Para un mayor rendimiento
dfMerge = dfMerge.head()

# Montamos nuestro dataframe principal
dfPrincipal = dfMerge[['movie_id','original_title', 'original_language', 'popularity', 
                       'release_date', 'revenue', 'runtime', 'status','vote_average', 
                       'vote_count']]
```

```python
# Montamos los dataframes para las listas

df = dfMerge[['movie_id', 'cast', 'crew', 'genres', 'keywords', 'production_countries',
             'spoken_languages']]

print("Longitud df: ", len(df))

# Inicializamos los dataframes vacíos
dfCast = []
dfCrew = []
dfGenres = []
dfKW = []
dfPCountries = []
dfSLanguages = []

# Nota: Por cada append, se vacía la variable que contiene el json para ahorrar memoria.
for i in range(len(df)):
    id = df['movie_id'][i]
    print("contador: ", i, " Id -> ", id )
    
    # Montamos el dataframe de casts
    castJson = pd.read_json(df['cast'][i])
    castJson['movie_id'] = id
    dfCast.append(castJson[['movie_id', 'name', 'character', 'gender']])
    castJson = [] 
    
    # Montamos el dataframe de crew
    crewJson = pd.read_json(df['crew'][i])
    crewJson['movie_id'] = id
    dfCrew.append(crewJson[['movie_id', 'name', 'gender', 'department', 'job']])
    crewJson = []
    
    # Montamos el dataframe de genres
    genresJson = pd.read_json(df['genres'][i])
    genresJson['movie_id'] = id
    dfGenres.append(genresJson[['movie_id', 'name']])
    genresJson = []
    
    # Montamos el dataframe de Keywords
    kwJson = pd.read_json(df['keywords'][i])
    kwJson['movie_id'] = id
    dfKW.append(kwJson[['movie_id', 'name']])
    kwJson = []
    
    # Montamos el dataframe de production_countries
    pcJson = pd.read_json(df['production_countries'][i])
    pcJson['movie_id'] = id
    dfPCountries.append(pcJson[['movie_id', 'name']])
    pcJson = []
    
    # Montamos el dataframe de spoken_languages
    slJson = pd.read_json(df['spoken_languages'][i])
    slJson['movie_id'] = id
    dfSLanguages.append(slJson[['movie_id', 'name']])
    slJson = []
    
    
# Concatenamos para que devuelva dataframe
dfCast = pd.concat(dfCast)
dfCrew = pd.concat(dfCrew)
dfGenres = pd.concat(dfGenres)
dfKW = pd.concat(dfKW)
dfPCountries = pd.concat(dfPCountries)
dfSLanguages = pd.concat(dfSLanguages)

print(dfCast.head())

# Vaciamos el dataframe dfMerge para un mejor rendimiento, no lo volveremos a usar.
dfMerge = []
```
