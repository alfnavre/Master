```python
def getPercentageOfProfitPerYear1(pd, dfMain):
    """
    @P1: pd --> pandas library
    @P2: dfMain --> main dataframe
    Returns --> % of profit of movies per year
    """
    df = dfMain.groupby(['release_year'])
    dfReturn = pd.DataFrame(columns=['release_year', 'sum_%_profit'])
    # List of all the years:
    years = list(df.groups.keys())
    cont = 0
    for year in years:
        # We obtain the movies. We filter by year and take into account there is no 0 values for revenue or budget.
        dfMovies = dfMain[(dfMain["release_year"] == year) &(dfMain['revenue'] != 0) & (dfMain['budget'] != 0)]
        # We set the year
        dfReturn.loc[cont, 'release_year'] = year
        # We set the profit per year
        profit = dfMovies["%Profitable"].reset_index()['%Profitable']
        # We set the profit for that year
        dfReturn.loc[cont,'sum_%_profit']= round(profit.sum(), 2)
        cont = cont + 1
        
    return(dfReturn)

df20 = getPercentageOfProfitPerYear1(pd, dfMain)
df20  
```

```python
""" 
    Dealing with Keybords Section: Bar Plot
    
    Objetive:  10 most popular keyboards of all movies in the dataframe
"""
dfKW = utl.getJSONDataFrame(pd, dfMerge, 'keywords')
# We aggrupate keybords by name, then sort them by count.
dfKWGrouped = dfKW.groupby(['name']).count().reset_index()
# Renaming the column.
dfKWGrouped = dfKWGrouped.rename(columns ={"movie_id": 'count'})
# Sorting the keybords
dfKWGrouped = dfKWGrouped.sort_values(by="count", ascending=False)
# Obtaining the first 20 more popular keybords
dfKWGrouped = dfKWGrouped.head(20)
# Convert the dataframe to a list which only contains the sorted Keybords.
keybordsList = dfKWGrouped['name'].tolist()
# Convert the dataframe to a list which only contains the sorted Counts.
countList = dfKWGrouped['count'].tolist()
# Obtaining the length of the objects list
y_pos = np.arange(len(keybordsList))

# Creating the plot
plt.barh(y_pos, countList, align='center', alpha=0.5)
plt.yticks(y_pos, keybordsList)
plt.xlabel('count')
plt.title('Most Popular Keyboards of all movies')
# Showing the plot
plt.show()

```

```python
def getDataFrameEconomicDataPerCountry(dfPCountries, dfMain, objeto):
    """
    @P1: dfPCountries --> dataframe containing the countries
    @P2 : dfMain --> main dataframe
    @P3: Objeto --> Campo a devolver budget | revenue
    Returns a dataframe that contains budget | revenue | runtime | popularity | vote_average 
    """
    dfReturn = pd.DataFrame(columns=['country',  objeto])
    cont = 1
    
    for i in range(1):
        # We get the id
        id = dfMain['movie_id'][i]
        # We get the data asked
        value = dfMain[objeto][i]
        print("cont: ", cont, " id: ", id, " value: ", value)
        # We get the countries for that id
        countriesPerMovie = dfPCountries[dfPCountries['movie_id'] == id]
        # Number of countries in the movie
        nCountriesPerMovie = len(countriesPerMovie)
        # We divide the value, so it is equal for all countries 
        value = value / nCountriesPerMovie
        print("new value: ", value)
        # We obtain the name of the countries 
        
        for country in countriesPerMovie['name'].tolist():
            print("for")
            if country in dfReturn['country'].tolist():
                print("if")
                # if the country is in the dataframe, we update it
                dfSelected = dfReturn.loc[dfReturn['country'] == country] 
                row = dfSelected.index.values[0]
                dfReturn.loc[row, objeto] = dfSelected[objeto].reset_index()[objeto][0] + value
                print("value for: ", country, " is: ", value)
            else:
                print("else")
                # if the country isnt in the dataframe, we add it
                dfReturn.loc[cont, 'country'] = country
                dfReturn.loc[cont, objeto] = value
                # we update cont here
                cont = cont +1
                
 
    return(dfReturn)

dfPCountries = utl.getJSONDataFrame(pd, dfMerge.head(20), 'production_countries')
df = getDataFrameEconomicDataPerCountry(dfPCountries, dfMain, 'budget')
print(df)
```
